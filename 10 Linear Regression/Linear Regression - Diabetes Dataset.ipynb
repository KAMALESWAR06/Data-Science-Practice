```python
import numpy as np
```

```python
train_data = np.genfromtxt('train.csv', delimiter = ',')
print(train_data)
print(train_data.shape)
```

```python
Y = train_data[:, -1]
Y
```

```python
X = train_data[:, :-1]
X
```

```python
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size = 0.3)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
ypred = model.predict(X_test)
```

```python
ypred
```

```python
def score(ypred, ytrue):
    u = ((ytrue - ypred) ** 2).sum()
    v = ((ytrue - ytrue.mean()) ** 2).sum()
    return 1 - u/v
```

```python
score(ypred, Y_test)
```

```python
Y_train_predicted = model.predict(X_train)
score(Y_train_predicted, Y_train)
```

```python
X_Test_Set = np.genfromtxt('test.csv', delimiter = ',')
X_Test_Set.shape
```

```python
Y_Test_predict = model.predict(X_Test_Set)
Y_Test_predict
```
