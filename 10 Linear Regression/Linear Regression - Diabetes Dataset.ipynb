```python
import numpy as np
```

```python
train_data = np.genfromtxt('train.csv', delimiter = ',')
print(train_data)
print(train_data.shape)
```

```python
Y = train_data[:, -1]
Y
```

```python
X = train_data[:, :-1]
X
```

```python
X.shape, Y.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
def score(ypred, ytrue):
    u = ((ytrue - ypred) ** 2).sum()
    v = ((ytrue - ytrue.mean()) ** 2).sum()
    return 1 - u/v
```

```python
Y_train_predicted = model.predict(X)
score(Y_train_predicted, Y)
```

```python
X_Test_Set = np.genfromtxt('test.csv', delimiter = ',')
X_Test_Set.shape
```

```python
Y_Test_predict = model.predict(X_Test_Set)
Y_Test_predict
```

```python
Y_Test_predict.shape
```
