### Plot the line graph between no. of Job postings with respect to year. Print the year and the number of job posting as integer value. Note: Year should be in ascending order.

```python
# Open and read data file as specified in the question
# Print the required output in given format
import numpy as np
import matplotlib.pyplot as plt
import csv
d = {}
with open('amazon_jobs_dataset.csv', 'r', encoding='utf8') as data:
    reader = csv.DictReader(data)
    for row in reader:
        d[int(row["Posting_date"][-4:])] = d.get(int(row["Posting_date"][-4:]), 0) + 1

x = np.arange(2011,2019,1)
y = []
for el in x:
    pass
    y.append(d[el])
y = np.array(y)
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Job Openings")
plt.title("Job Openings per year")
plt.grid()
plt.show()
for i in range(len(x)):
    print(x[i], y[i])
```

### Plot the Bar graph between Month vs Job Openings. Print the year and the number of job posting as integer value. Order of months doesn't matter.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
d = {}
with open('amazon_jobs_dataset.csv', 'r', encoding='utf8') as data:
    reader = csv.DictReader(data)
    for row in reader:
        month = row["Posting_date"].split()[0]
        d[month] = d.get(month, 0) + 1

x = np.array(["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
y = []
for el in x:
    y.append(d[el])
y = np.array(y)
plt.bar(x,y)
plt.xlabel("Year")
plt.ylabel("Job Openings")
plt.title("Job Openings per year")
plt.grid()
plt.xticks(rotation = 45)
plt.show()
for i in range(len(x)):
    print(x[i], y[i])
```

### Plot the scatter graph between year vs No. of jobs opening related to Java. Print the year and number of Jobs opening in Java Profile. Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
d = {}
with open('amazon_jobs_dataset.csv', 'r', encoding='utf8') as data:
    reader = csv.DictReader(data)
    for row in reader:
        if "Java" in row["BASIC QUALIFICATIONS"] or "java" in row["BASIC QUALIFICATIONS"]:
            d[int(row["Posting_date"][-4:])] = d.get(int(row["Posting_date"][-4:]), 0) + 1
l1 = []
l2 = []
for el in d:
    l1.append(el)
    l2.append(d[el])
plt.scatter(l1,l2)
plt.show()
for i in range(len(l1)-1,-1,-1):
    print(l1[i], l2[i])
```

### Plot the Pie chart between Indian cities vs No. of jobs opening. Print the Indian cities and %age of Job distribution in India upto 2 decimal places. Note: %age of Job distribution should be in descending order.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
df = pd.read_csv('amazon_jobs_dataset.csv', encoding='utf8')
loc = df["location"]
loc = loc.str.split(',', expand = True)
loc = loc[loc[0] == "IN"]
city = loc[2]
np_cities = np.array(city.value_counts().index)
np_jobs = np.array(city.value_counts().values)
plt.pie(np_jobs, labels = np_cities, autopct = '%.2f%%')
plt.show()

for i in range(city.nunique()):
    print(city.value_counts().index[i], end = ' ')
    print(city.value_counts().values[i] * 100 / city.count())
```
