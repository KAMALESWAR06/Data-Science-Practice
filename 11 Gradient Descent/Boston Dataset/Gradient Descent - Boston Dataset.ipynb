```python
import numpy as np
```

```python
data = np.genfromtxt('training_boston_x_y_train.csv', delimiter = ',')
x = data[:,:-1]
y = data[:, -1]


from sklearn import preprocessing
sc = preprocessing.StandardScaler()
sc.fit(x)
x = sc.transform(x)

x,y,x.shape,y.shape
```

```python
test_data = np.genfromtxt('test_boston_x_test.csv', delimiter = ',')
test_data = sc.transform(test_data)
```

```python
def cost(x,y,m):
    final_cost = 0
    r = x.shape[0]
    c = x.shape[1]
    for i in range(r):
        ypred = 0
        for j in range(c):
            ypred += m[j] * x[i][j]
        final_cost += (1/r) * ((y[i] - ypred) ** 2)
    print(final_cost)
    
def gradientStep(x,y,coeff,alpha):
    rows = x.shape[0]
    features = x.shape[1]
    slope = [0 for i in range(features)]
    for i in range(rows):
        m_x = 0
        for j in range(features):
            m_x += coeff[j] * x[i,j]
        for j in range(features):
            slope[j] += (-2/rows) * (y[i] - m_x) * x[i][j]
    for j in range(features):
        coeff[j] -= alpha * slope[j]
    
def gradientDescent(x,y,alpha,itr):
    x = np.append(x, np.ones((x.shape[0], 1)), axis = 1)
    rows = x.shape[0]
    features = x.shape[1]
    coeff = [0 for i in range(features)]
    for _itr in range(itr):
        gradientStep(x,y,coeff,alpha)
        cost(x,y,coeff)
    return coeff
```

```python
m = gradientDescent(x,y,0.1,500)
m
```

```python
def predict(x,m):
    y = []
    for el in x:
        y_pred = 0
        for j in range(len(el)):
            y_pred += m[j] * el[j]
        y.append(y_pred + m[-1])
    return np.array(y, dtype = float)
print(m)
```

```python
test_y_predicted = predict(test_data,m)
print(test_y_predicted)
```
