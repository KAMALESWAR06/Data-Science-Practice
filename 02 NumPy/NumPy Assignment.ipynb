### Find value of killed column only where country == ‘United States’? Print 0 in place of missing values. Print count of Killed as integer value.

```python
import numpy as np
import csv

with open('datasets/terrorismData.csv', 'r', encoding='utf8') as data:
    reader=csv.DictReader(data) 
    killed=[] 
    for row in reader: 
        if row['Country']=='United States': 
            killed.append(row['Killed'])   
npkilled=np.array(killed) 
npkilled[npkilled==''] = '0.0'
npkilled=np.array(npkilled,dtype=float) 
for el in npkilled:
    print(int(el))
```

### Find the number of attack held between 1 Jan 2010 and 31 Jan 2010?(including both date). Note Ignore the case where day is 0. Print count of NumberOFAttack as integer value.

```python
import numpy as np
import csv
killed = []
wounded = []
city = []
group = []

with open('datasets/terrorismData.csv', 'r', encoding = 'utf8') as data:
    reader = csv.DictReader(data)
    for row in reader:
        if row["State"] == "Jammu and Kashmir" and int(row["Year"]) == 1999 and int(row["Month"]) in [5,6,7]:
            killed.append(row["Killed"])
            wounded.append(row["Wounded"])
            city.append(row["City"])
            group.append(row["Group"])

npk = np.array(killed)
npw = np.array(wounded)
npc = np.array(city)
npg = np.array(group)
npw[npw == ''] = 0.0
npk[npk == ''] = 0.0
npw = np.array(npw, dtype = float)
npk = np.array(npk, dtype = float)
npcasulties = npw + npk
idx = npcasulties.argmax()
print(int(npcasulties[idx]), npc[idx], npg[idx])
```

### Find the casualty in the Red Corridor States ? Mainly Red corridor states include Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh.

```python
import numpy as np
import csv

killed = []
wounded = []
with open('datasets/terrorismData.csv', 'r', encoding = 'utf8') as data:
    reader = csv.DictReader(data, skipinitialspace = True)
    for row in reader:
        if row["State"] in ["Jharkhand", "Odisha", "Andhra Pradesh", "Chhattisgarh"]:
            killed.append(row["Killed"])
            wounded.append(row["Wounded"])
np_k = np.array(killed)
np_w = np.array(wounded)
np_k[np_k == ''] = 0
np_w[np_w == ''] = 0
np_k = np.array(np_k, dtype = float)
np_w = np.array(np_w, dtype = float)
print(int(np.sum(np_k) + np.sum(np_w)))
```

### Find top 5 Indian Cities which has most number of casualties ? Print top 5 cities along with total casualties in that city. Print count of Casualty as integer value. Note: Ignoring the City which is Unknown. Casualty = Killed + Wounded.

```python
import numpy as np
import csv
import heapq
casualties = []
city = []
with open('datasets/terrorismData.csv', 'r', encoding = 'utf8') as data:
    reader = csv.DictReader(data, skipinitialspace = True)
    for row in reader:
        if row["Country"] == "India" and row["City"] != "Unknown":
            casualties.append([row["Killed"], row["Wounded"]])
            city.append(row["City"])
np_cas = np.array(casualties)
np_cas[np_cas == ''] = 0
np_cas = np.array(np_cas, dtype = float)
c_d = {}
for i in range(len(city)):
    c_d[city[i]] = c_d.get(city[i], 0) + int(np_cas[i][0] + np_cas[i][1])
minheap = []
for el in c_d:
    if len(minheap) < 5:
        heapq.heappush(minheap, (c_d[el], el))
    elif len(minheap) == 5:
        if minheap[0][0] < c_d[el]:
            heapq.heapreplace(minheap, (c_d[el], el))
ans = []
while len(minheap) > 0:
    el = heapq.heappop(minheap)
    ans.append(el)
ans = ans[::-1]
for el in ans:
    print(el[1], el[0])
```

### Find the most frequent day of attack in a terrorismDataset ? Note: Here np.unique can be used. Print count of frequent day and number of attack as Integer value.
